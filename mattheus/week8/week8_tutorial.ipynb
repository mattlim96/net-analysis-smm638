## Import packages

```python
import numpy as np
import networkx as nx
from pprint import pprint as pp
import pandas as pd
import random
```

## Simulate a network and node attribute

```python
g = nx.Graph()
seed = 1
n = 200
# simulate a network with a random graph
g.add_edges_from(nx.gnm_random_graph(n, 50, seed=seed).edges())

for node in g.nodes:
    # random 0 and 1 
    g.nodes[node]['attr'] = random.randint(0, 1)
```

## Plot data

```python
# draw the network
pos = nx.layout.kamada_kawai_layout(g)
colors = []
for n in g.nodes():
    if g.nodes[n]['attr'] == 1:
        colors.append('red')
    else:
        colors.append('blue')
nx.draw(g, pos=pos, with_labels=False, node_color=colors)
```
